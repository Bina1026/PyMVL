# PyMVL Tutorial

## Overview

This tutorial will introduce you to the basics of PyMVL, a library designed for Creating Heatmaps of Radiomics Analysis. You will learn how to install it, use its core functionalities, and see examples to help you get started quickly.

## Library Import

```python
import pickle
import numpy as np
import pyMVL
import os
```

```python
base_dir = 'Sample'

global_feature_csv = os.path.join(base_dir, 'total_datasets.csv')
shap_feature_csv = os.path.join(base_dir, 'total_shap.csv')
img_path = os.path.join(base_dir, 'images.pkl')
mask_path = os.path.join(base_dir, 'masks.pkl')
```

### You can easily pick how many numbers you want to visualize (in this case = 5)

```python
extractor = pyMVL.Extractor(global_feature_csv, shap_feature_csv, top_no=5)
```

```python
extractor.visualize_interpolation()
```

```python
with open(img_path, 'rb') as file:
    images = pickle.load(file)

with open(mask_path, 'rb') as file:
    masks = pickle.load(file)
```

```python
kys = list(images.keys())
selected_kys = np.random.choice(kys, int(0.03*len(kys)), replace=False)

images_selected = {key : images[key] for key in selected_kys}
masks_selected = {key : masks[key] for key in selected_kys}
```

```python
params = {}

params['binWidth'] = 5
```

```python
extractor.execute(images=images_selected, masks=masks_selected, kernel_size=4, label=255, params=params)
```
